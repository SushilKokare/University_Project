```python
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array, array_to_img
```

```python
datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=60,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='nearest'

)
```

```python
img = load_img('micky_mouse.jpg')
```

```python
img
```

```python
x = img_to_array(img)
```

```python
x.shape
```

```python
x = x.reshape((1,)+x.shape)
```

```python
x.shape
```

```python
i = 0
for batch in datagen.flow(x, batch_size=1,
                         save_to_dir='Aug_Data', save_prefix='mm',
                         save_format='jpg'):
    i += 1
    if i > 20:
        break
```

```python
import os
```

```python
img_folder = 'Brain_Data_Organised/Train/Stroke/'
```

```python
img_files = os.listdir(img_folder)
```

```python
import time
```

```python
start_time = time.time()
for img_file in img_files:
    img_path = os.path.join(img_folder,img_file)
    img = load_img(img_path)
    
    x = img_to_array(img)
    x = x.reshape((1,)+x.shape)
    i = 0
    for batch in datagen.flow(x, batch_size=1,
                         save_to_dir='Aug_Data_MRI_Train_Stroke', save_prefix='admri',
                         save_format='jpg'):
        i += 1
        if i > 2:
            break
end_time = time.time()
```
